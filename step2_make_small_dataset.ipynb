```python
import torch
import pickle
import pandas as pd
pd.set_option('display.min_rows', 4)
```

```python
# select only a subset of the dataset

M        = 1000     # number of movies
min_freq = 10       # drop users who watched less than this

df = (pd.read_csv('ratings.csv')
      .query(f'movie < {M}')
      .groupby('user').filter(lambda x: len(x)>=min_freq)
      )
df
```

```python
# reindexing users to consecutive integers (from 0 to N)

old_user_ids = df.sort_values('user').user.unique()
new_id_map = dict(zip(old_user_ids, range(len(old_user_ids))))
df = (df
      .assign(user=df.user.map(new_id_map))
      .sort_values(['movie', 'user'])
      .reset_index(drop=True)
     )
N = df.user.max()

print('num users:', N)
df
```

```python
# save new dataset
df.to_csv(f'ratings_M{M}-N{N}.csv', index=False)
```
